```python
pickle_path = r'/home/sandro/Projekte/github_projects/deeptech-ai/data/raw/companies.pickle'
```

```python
errors_file = r'/home/sandro/Projekte/github_projects/deeptech-ai/data/raw/errors.txt'
```

```python
errors_urls = []

with open(errors_file, 'r') as efile:
    lines = efile.readlines()
    for line in lines:
        line = line.replace('\n', '')
        line = line.split('. ')
        line = line[0]
        line = line.replace(r'\n', '')
        if line[-1] == '.':
            line = line[:-1]
        errors_urls.append(line)
```

```python
errors_urls
```

```python
import pickle
favorite_color = pickle.load( open( pickle_path, "rb" ) )
```

```python
df = favorite_color
```

```python
a = [ d in errors_urls for d in df['url']]
```

```python
df[a]
```

```python
dfa = df[a]
```

```python
import numpy as np
```

```python
k = dfa.groupby('section').size() / np.sum(dfa.groupby('section').size())
```

```python
from matplotlib import pyplot as plt
```
